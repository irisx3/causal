```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LogisticRegression

X = total_data_long_corrected[['unem', 'log_gdp']]
T = total_data_long_corrected['treatment']
Y = total_data_long_corrected['log_apps']

# Estimating Propensity Scores 
propensity_model = LogisticRegression()
propensity_model.fit(X, T)
propensity_scores = propensity_model.predict_proba(X)[:, 1]

# Outcome Regression using Random Forest
outcome_model = RandomForestRegressor(random_state=42)
outcome_model.fit(X.join(pd.DataFrame(T)), Y)

# Calculating Doubly Robust Estimator 
def doubly_robust_estimate(T, Y, X, propensity_scores, outcome_model):
    N = len(Y)
    dr_estimates = []

    for i in range(N):
        X_with_treatment = X.iloc[[i]].assign(treatment=1)
        X_with_untreatment = X.iloc[[i]].assign(treatment=0)

        predicted_outcome_treated = outcome_model.predict(X_with_treatment)
        predicted_outcome_untreated = outcome_model.predict(X_with_untreatment)

        weight_treated = T.iloc[i] / propensity_scores[i]
        weight_untreated = (1 - T.iloc[i]) / (1 - propensity_scores[i])

        dr = (weight_treated * (Y.iloc[i] - predicted_outcome_treated)) + \
             (weight_untreated * (Y.iloc[i] - predicted_outcome_untreated)) + \
             predicted_outcome_treated - predicted_outcome_untreated

        dr_estimates.append(dr)

    return np.mean(dr_estimates)

dr_estimate = doubly_robust_estimate(T, Y, X, propensity_scores, outcome_model)
dr_estimate

```

```python
from sklearn.ensemble import RandomForestRegressor
import numpy as np


years = ['2012', '2013', '2014', '2015', '2016', '2017']
data_long = pd.DataFrame()

for year in years:
    temp = data[['Country', year, f'weight_{year}', f'{year}_unem', f'{year}_gdp']].copy()
    temp.rename(columns={year: 'applicants', f'weight_{year}': 'weight', f'{year}_unem': 'unem', f'{year}_gdp': 'gdp'}, inplace=True)
    temp['year'] = int(year)
    data_long = pd.concat([data_long, temp])


data_long = data_long[data_long['Country'] != 'Total']


data_long['treatment'] = data_long['year'].apply(lambda x: 0 if x < 2014 else 1)

# Log transformation of applicants and GDP
data_long['log_apps'] = np.log(data_long['applicants'].clip(lower=1))  # Using clip to avoid log(0)
data_long['log_gdp'] = np.log(data_long['gdp'].clip(lower=1))           # Using clip to avoid log(0)


data_long
```

```python

X = pd.get_dummies(data_long[['Country', 'unem', 'log_gdp']], drop_first=False)
X['treatment'] = data_long['treatment']


Y = data_long['log_apps']


rf_model = RandomForestRegressor(random_state=55)
rf_model.fit(X, Y)


cate_estimates = {}
for country in data_long['Country'].unique():
    country_column = f'Country_{country}'
    
    
    if country_column not in X.columns:
        print(f"Country '{country}' not found in the encoded DataFrame.")
        continue

    country_mask = X[country_column] == 1

    
    if X[country_mask].empty or len(X[country_mask]['treatment'].unique()) < 2:
        print(f"Insufficient data for country: {country}")
        continue

    X_country_treated = X[country_mask].copy()
    X_country_treated['treatment'] = 1
    X_country_untreated = X[country_mask].copy()
    X_country_untreated['treatment'] = 0

    
    predicted_outcome_treated = rf_model.predict(X_country_treated).mean()
    predicted_outcome_untreated = rf_model.predict(X_country_untreated).mean()
    
    # CATE estimation for the country
    cate_estimates[country] = predicted_outcome_treated - predicted_outcome_untreated

cate_estimates



```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression, LinearRegression
import pandas as pd
import numpy as np

Dn1, Dn2 = train_test_split(data_long, test_size=0.5, random_state=42)


def dr_learner_steps(Dn_train, Dn_test):
    
    propensity_model = LogisticRegression()
    propensity_model.fit(Dn_train[['unem', 'log_gdp']], Dn_train['treatment'])
    b_pi = propensity_model.predict_proba(Dn_train[['unem', 'log_gdp']])[:, 1]

    
    outcome_model_treated = LinearRegression()
    outcome_model_untreated = LinearRegression()

    outcome_model_treated.fit(Dn_train[Dn_train['treatment'] == 1][['unem', 'log_gdp']], Dn_train[Dn_train['treatment'] == 1]['log_apps'])
    outcome_model_untreated.fit(Dn_train[Dn_train['treatment'] == 0][['unem', 'log_gdp']], Dn_train[Dn_train['treatment'] == 0]['log_apps'])

    b_mu1 = outcome_model_treated.predict(Dn_train[['unem', 'log_gdp']])
    b_mu0 = outcome_model_untreated.predict(Dn_train[['unem', 'log_gdp']])

    # Pseudo-outcome Regression
    Dn_train['pseudo_outcome'] = (Dn_train['treatment'] - b_pi) / (b_pi * (1 - b_pi)) * (Dn_train['log_apps'] - Dn_train['treatment'] * b_mu1 - (1 - Dn_train['treatment']) * b_mu0) + b_mu1 - b_mu0
    pseudo_outcome_model = LinearRegression()
    pseudo_outcome_model.fit(Dn_train[['unem', 'log_gdp']], Dn_train['pseudo_outcome'])

    return pseudo_outcome_model.predict(Dn_test[['unem', 'log_gdp']])


tau_dr_1 = dr_learner_steps(Dn1, Dn2)
tau_dr_2 = dr_learner_steps(Dn2, Dn1)


final_cate_estimates = {}
for dataset in [Dn1, Dn2]:
    for i, row in dataset.iterrows():
        country = row['Country']
        tau_dr = tau_dr_1[i] if i in Dn1.index else tau_dr_2[i - len(Dn1)]
        if country in final_cate_estimates:
            final_cate_estimates[country].append(tau_dr)
        else:
            final_cate_estimates[country] = [tau_dr]


for country in final_cate_estimates:
    final_cate_estimates[country] = np.mean(final_cate_estimates[country])

final_cate_estimates

```

```python
from sklearn.utils import resample
import numpy as np


n_bootstraps = 1000
bootstrap_estimates = {country: [] for country in data_long['Country'].unique()}

for _ in range(n_bootstraps):
   
    bootstrap_sample = resample(data_long)

    
    tau_dr_bootstrap = dr_learner_steps(bootstrap_sample, bootstrap_sample)


    country_bootstrap_estimates = {}
    for i, row in bootstrap_sample.iterrows():
        country = row['Country']
        if country in country_bootstrap_estimates:
            country_bootstrap_estimates[country].append(tau_dr_bootstrap[i])
        else:
            country_bootstrap_estimates[country] = [tau_dr_bootstrap[i]]


    for country in country_bootstrap_estimates:
        avg_ate = np.mean(country_bootstrap_estimates[country])
        bootstrap_estimates[country].append(avg_ate)


ci_estimates = {}
for country, estimates in bootstrap_estimates.items():
    ci_lower = np.percentile(estimates, 2.5)
    ci_upper = np.percentile(estimates, 97.5)
    ci_estimates[country] = (ci_lower, ci_upper)

ci_estimates
```
